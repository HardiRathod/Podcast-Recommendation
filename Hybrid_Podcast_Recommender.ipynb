```python
df_dup = df[['title', 'podcast_id','genre_tags', 'desc', 'url', 'tf-idf']]
df_dup.drop_duplicates(inplace=True)
df_dup
```

```python
df_star = pd.DataFrame(df.groupby('podcast_id')['stars'].mean())
df_star
```

```python
merge = df_dup.merge(df_star, left_on='podcast_id', right_on='podcast_id')
merge['genre_tags'] = merge['genre_tags'].map(lambda i: i.split(','))
```

```python
s = merge.apply(lambda x: pd.Series(x['genre_tags']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'genre'
new_merge = merge.drop('genre_tags', axis=1).join(s)
new_merge.head(3).transpose()
```

```python
#getting comedy podcasts
#Sorted by stars
df_temp = new_merge[new_merge['genre'] == "Comedy"]

incl = df_temp[(df_temp['stars'].notnull())][['title', 'desc', 'url','stars','genre']]
incl = incl.sort_values('stars', ascending=False).head(250)
```

```python
# Understanding Podcasts transcripts and descriptions
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfid_matrix = tf.fit_transform(merge['desc'])
```

```python
tfid_matrix
```

```python
#Cosine Similarity
cossim = linear_kernel(tfid_matrix, tfid_matrix)
```

```python
cossim
```

```python
merge = merge.reset_index(drop=True)
titles = merge[['title', 'genre_tags', 'stars']]
indices = pd.Series(merge.index, index=merge['title'])
indices.head(5)
```

```python
#Finding Similar Recommendations
def find_recom(title):
    idx = indices[title]
    sim_scores = list(enumerate(cossim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    #Adjust this
    sim_scores = sim_scores[1:41]
    podcast_indices = [i[0] for i in sim_scores]
    return titles.iloc[podcast_indices]
find_recom('Over My Dead Body')
```

```python
# Finding recommendations with Content Based Recommender
```

```python
merge['desc_keywords'] = merge['desc'].apply(lambda x: x.split(' '))
merge['genre_tags_new'] = merge['genre_tags'].apply(lambda x: x*3)
merge
```

```python
merge['desc_genre'] = merge['desc_keywords'] + merge['genre_tags_new']
merge['desc_genre'] = merge['desc_genre'].apply(lambda x: ' '.join(x))
```

```python
count = CountVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
count_matrix = count.fit_transform(merge['desc_genre'])
cossim = cosine_similarity(count_matrix, count_matrix)
```

```python
merge = merge.reset_index(drop=True)
titles =  merge[['title', 'genre_tags', 'stars']]
indices = pd.Series(merge.index, index=merge['title'])
```

```python
find_recom('Over My Dead Body')
```

```python
#recommendations by rating
def recom_by_rating(title):
    idx = indices[title]
    sim_scores = list(enumerate(cossim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:26]
    podcast_indices = [i[0] for i in sim_scores]
    
    podcast = merge.iloc[podcast_indices][['title', 'desc', 'url','stars','genre_tags']]
    podcast = podcast.sort_values('stars', ascending=False).head(10)
    return podcast
recom_by_rating('Unsolved Murders')
```
